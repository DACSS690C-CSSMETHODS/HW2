Following people on social media is a traditional example of an asymmetric relationship.

These data is on another Excel on GitHub:

```python
# link to GitHub
linkGoogle="https://github.com/DACSS-CSSmeths/Networks_intro/raw/refs/heads/main/casesData/SeattleTopData.xlsx"
```

Turning this data frame into a graph:

```python
edgelist = pd.read_excel(linkGoogle,sheet_name='edges')

edgelist
```

```python
#Networkx can creat a network if your data frame has those columns names (source and target):
graph_edgelist=nx.from_pandas_edgelist(edgelist, edge_attr=True,
                                         create_using=nx.DiGraph)
```

```python
# see edges
graph_edgelist.edges.data()
```

And turning this dataframe into a graph:

```python
attr = pd.read_excel(linkGoogle,
                          sheet_name='attributes')

attr
```

```python
# dictionary for attributes
attrDic_male={n:s for n,s in zip(attr.name,attr.male)}
attrDic_followers={n:s for n,s in zip(attr.name,attr.followers)}
nx.set_node_attributes(graph_edgelist,attrDic_followers , "followers")
nx.set_node_attributes(graph_edgelist,attrDic_male , "male")
```

```python
graph_edgelist.nodes.data()
```

```python
graph_edgelist.edges.data()
```

Prepare the right network, including the node attributes, and save (export) the network as a graphml file.

```python
# use the edge attributes
# add labels to edges

pos = nx.spring_layout(graph_edgelist) # position of the nodes

nx.draw(graph_edgelist,
        pos, # using "position"
        with_labels=True,
        node_color=nx.get_node_attributes(graph_edgelist, "male").values())

# adding labels
final_dG=nx.draw_networkx_edge_labels(graph_edgelist,pos,edge_labels=nx.get_edge_attributes(graph_edgelist,'weight'))
```

### Saving my graph

```python
myFolder = "MyGraphs/"
nx.write_graphml(graph_edgelist, myFolder + "graph_edgelist.graphml", named_key_ids=True)
```
