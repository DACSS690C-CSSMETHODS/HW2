### Goal: Prepare the right (directed or undirected?) network, including the node attributes, and save (export) the network as a graphml file.

```python
#now lets turn the first dataframe into a graph
graph_FigueData = nx.from_pandas_adjacency(FigueData, create_using=nx.DiGraph)
#see edges
graph_FigueData.edges.data()
```

```python
#i want to add the node attributes to graph_FigueData
graph_FigueData.nodes.data()
```

```python
#I am building a dictionary of attributes
{n:s for n,s in zip(FigueAttr.node,FigueAttr.Multinacional)}
```

```python
#looked good above, following through with it
FigueAttrDic_Multinacional={n:s for n,s in zip(FigueAttr.node,FigueAttr.Multinacional)}
nx.set_node_attributes(graph_FigueData,FigueAttrDic_Multinacional , "Multinacional")
```

```python
#check node attributes now
graph_FigueData.nodes.data()
```

```python
#i want to prepare an undirected network on FigueData
pos = nx.circular_layout(graph_FigueData)
#draw nodes first
nx.draw_networkx_nodes(graph_FigueData,pos,node_color = nx.get_node_attributes(graph_FigueData,'Multinacional').values())
#draw edges
nx.draw_networkx_edges(graph_FigueData,pos,
                       width=list(nx.get_edge_attributes(graph_FigueData,'weight').values()))
#draw node labels
nx.draw_networkx_labels(graph_FigueData, pos)

nx.draw_networkx_edge_labels(graph_FigueData,pos,font_size=8, label_pos=0.25,
                               edge_labels=nx.get_edge_attributes(graph_FigueData,'weight'));
```

```python
# simpler graph
nx.draw_circular(graph_FigueData,with_labels=True,node_color='white',edgecolors='black')
```

### Saving my graph

```python
myFolder = "MyGraphs/"
nx.write_graphml(graph_FigueData, myFolder + "graph_FigueData.graphml", named_key_ids=True)
```
